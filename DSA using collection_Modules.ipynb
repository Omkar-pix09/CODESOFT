```
from collections import OrderedDict
```

```
od= OrderedDict()
#dir(od)
od.fromkeys("omkar")
```

```
# create a new Ordered Dictionary froom keys

od = {}.fromkeys("omkar",0)
print(od)
```

```
#Move the first character to the last
new_dict =OrderedDict(od)
new_dict
```

```
new_dict.move_to_end("o")
new_dict
```

```
new_dict.move_to_end("o")
new_dict.move_to_end("a")
new_dict
```

```
#pop the item in new_dict
new_dict.move_to_end("o")
new_dict.move_to_end("a")
new_dict.popitem()
print(new_dict)
```

```
new_dict.popitem(last=False)
print(new_dict)
```

```
print("".join(new_dict))
```

# Counter

```
from collections import Counter
```

```
freq= Counter("mississppi")
print(freq)
```

```
# Top 3 Highest Occuring element
freq.most_common(3)
```
